```python
# Median Spatial Domain Filtering


import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the image 
noisy_img = cv2.imread(r"C:\Users\ACER\Desktop\myProj\what-I-learnt-as-a-Computer-Engineering-Student\BE Computer\Semester\Sem-VII\Image-Processing-n-Pattern-Recognition\IPPR-Lab\Histogram\noisy-star.jpg", 0)

# rose = cv2.resize(rose, (800,600))  # Resizing the image

# obtain the number of rows and columns of the image
m,n = noisy_img.shape
print(m,n)

plt.imshow(noisy_img)
plt.set_cmap('gray')
```

```python
# Traverse the image. For every 3x3 area, 
# find the median of the pixels and replace the center pixel by the median

img_newm = np.zeros([m, n])

for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [noisy_img[i-1, j-1],
        noisy_img[i-1, j],
        noisy_img[i-1, j+1],
        noisy_img[i, j-1],
        noisy_img[i, j],
        noisy_img[i, j+1],
        noisy_img[i+1, j-1],
        noisy_img[i+1, j],
        noisy_img[i+1, j+1]]

        temp = sorted(temp)
        img_newm[i, j] = temp[4]

img_newm = img_newm.astype(np.uint8)
cv2.imwrite('new_median_filtered.png', img_newm)

```

```python
# show the original image
plt.subplot(121)
plt.imshow(noisy_img)
plt.title('original image')
plt.set_cmap('gray')

# show filtered image
plt.subplot(122)
plt.imshow(img_newm)
plt.title('with median filter')
plt.set_cmap('gray')
```
