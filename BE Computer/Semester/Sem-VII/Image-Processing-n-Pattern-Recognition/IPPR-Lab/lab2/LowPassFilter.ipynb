```python
# Low Pass Spatial Domain Filtering to obe=serve the blurring effect

from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the image 
# rose = cv2.imread(r"C:\Users\ACER\Desktop\myProj\what-I-learnt-as-a-Computer-Engineering-Student\BE Computer\Semester\Sem-VII\Image-Processing-n-Pattern-Recognition\IPPR-Lab\Histogram\Day1\pictures\rose.jpg", 0)
noisy_img = cv2.imread(r"C:\Users\ACER\Desktop\myProj\what-I-learnt-as-a-Computer-Engineering-Student\BE Computer\Semester\Sem-VII\Image-Processing-n-Pattern-Recognition\IPPR-Lab\Histogram\noisy-star.jpg", 0)

# rose = cv2.resize(rose, (800,600))  # Resizing the image

# obtain the number of rows and columns of the image
m,n = noisy_img.shape
print(m,n)

plt.imshow(noisy_img)
plt.set_cmap('gray')
```

```python
# function to obtain histogram of an image
def filter(img,sz):
    # Develop Averaging filter(sz, sz) mask
    mask = np.ones([sz, sz], dtype = int)
    mask = mask / (sz*sz)
    print(mask)
    # Convolce the 3x3 mask over the image
    img_new = np.zeros([m, n])
    for i in range(1, m-1):
        for j in range(1, n-1):

            temp = img[i-1, j-1]* mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j+1]*mask[0, 2]+img[i, j]*mask[1, 1]+img[i, j+1]*mask[1,2]+img[i+1, j-1]*mask[2,0] + img[i+1, j]*mask[2, 1] + img[i+1, j+1]*mask[2, 2]

            img_new[i, j] = temp

    img_new = img_new.astype(np.uint8)
    return img_new
```

```python
# show original image
plt.subplot(121)
plt.imshow(noisy_img)
plt.title('original image')
plt.set_cmap('gray')

# show filtered image
plt.subplot(122)
img_new3 = filter(noisy_img, 3)
plt.imshow(img_new3)
plt.title('with filter 3*3')
plt.set_cmap('gray')
```

```python
img_new5 = filter(noisy_img, 5)
img_new9 = filter(noisy_img, 9)
```

```python
plt.subplot(121)

plt.imshow(img_new5, cmap='gray')
plt.title('with filter 5*5')

plt.subplot(122)
plt.imshow(img_new9, cmap='gray')
plt.title('with filter 9*9')
```

```python
img_new15 = filter(noisy_img, 15)
img_new35 = filter(noisy_img, 35)
```

```python
plt.subplot(121)
plt.imshow(img_new15)
plt.title('with filter 15*15')
plt.set_cmap('gray')

plt.subplot(122)
plt.imshow(img_new35)
plt.title('with filter 35*35')
plt.set_cmap('gray')
plt.show()
```
