```python
def user_def_hist_plot(img):

    # variable to store intensity values
    intensity_count = []
    intensity_range = []

    # Initializing intensity count to a series of 256 0 values
    for c in range(256):
        intensity_count.append(0)
        intensity_range.append(c)

    m,n = img.shape

    for i in range(m):
        for j in range(n):
            intensity_count[img[i, j]] += 1

    return (intensity_range, intensity_count)
    
```

```python
intensity_range_var, intensity_count_var = user_def_hist_plot(img)
```

```python
# Histogram Stretching using builtin function

# import cv2, numpy, matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt

Image = 'Sophia Robot.jpg'
# Image = 'einstein.jpg'
img = cv2.imread(Image, 0)
equ = cv2.equalizeHist(img)
res = np.hstack((img, equ))  # stacking image side-by-side
cv2.imwrite('res.png', res)
plt.imshow(res)
```
